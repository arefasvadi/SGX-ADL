# Clean Runs of Datasets

## CIFAR-10 First Network

```python
import numpy as np
import matplotlib.pyplot as plt
import csv

plt.figure(num=None, figsize=(10, 7.5), dpi=80, facecolor='w', edgecolor='k')

epochs = np.arange(1,101)
#print epochs

train_acc_mod1 = []
test_acc_mod1 = []

with open('saved_models/cifar10-1/run-log-temp.txt', 'rb') as f:
    reader = csv.reader(f)
    for row in reader:
        train_acc_mod1.append(float(row[0]))
        test_acc_mod1.append(float(row[1]))
        
#print test_acc
plt.clf()
#plt.yticks(np.arange(0,101))
plt.plot(epochs,train_acc_mod1,label='train')
plt.plot(epochs,test_acc_mod1,label='test')
plt.legend(loc='lower right', frameon=False)
plt.title("Network 1")
plt.grid(linestyle = '-.')
plt.show()
```

# Noisy Weights

## Cifar10 first Network with proportions in range {0.2, 0.1, 0.05, 0.005, 0.0001} and  

## Standard Deviation in range {5x10^(-4), 10^(-4), 10 ^ (-5), 10 ^ (-6)}

```python
import numpy as np
import matplotlib.pyplot as plt
import csv

plt.figure(num=None, figsize=(10, 7.5), dpi=80, facecolor='w', edgecolor='k')

epochs = np.arange(1,101)
#print epochs

test_acc = []
test_acc_noise_1 = [] # PROPORTION: 0.2, STD = 0.005
test_acc_noise_2 = [] # PROPORTION: 0.1, STD = 0.005
test_acc_noise_3 = [] # PROPORTION: 0.05, STD = 0.005
test_acc_noise_4 = [] # PROPORTION: 0.005, STD = 0.005
test_acc_noise_5 = [] # PROPORTION: 0.0005, STD = 0.005


with open('saved_models/cifar10-1/run-log-temp.txt', 'rb') as f:
    read_len = 0
    reader = csv.reader(f)
    for row in reader:
        if read_len < len(epochs):
            test_acc.append(float(row[1]))
        read_len =read_len + 1

with open('saved_models/cifar10-1-noise-0.2-0.0005/run-log-temp.txt', 'rb') as f:
    read_len = 0
    reader = csv.reader(f)
    for row in reader:
        if read_len < len(epochs):
            test_acc_noise_1.append(float(row[0]))
        read_len =read_len + 1

with open('saved_models/cifar10-1-noise-0.1-0.0005/run-log-temp.txt', 'rb') as f:
    read_len = 0
    reader = csv.reader(f)
    for row in reader:
        if read_len < len(epochs):
            test_acc_noise_2.append(float(row[0]))
        read_len =read_len + 1
        
with open('saved_models/cifar10-1-noise-0.05-0.0005/run-log-temp.txt', 'rb') as f:
    read_len = 0
    reader = csv.reader(f)
    for row in reader:
        if read_len < len(epochs):
            test_acc_noise_3.append(float(row[0]))
        read_len =read_len + 1

with open('saved_models/cifar10-1-noise-0.005-0.0005/run-log-temp.txt', 'rb') as f:
    read_len = 0
    reader = csv.reader(f)
    for row in reader:
        if read_len < len(epochs):
            test_acc_noise_4.append(float(row[0]))
        read_len =read_len + 1

with open('saved_models/cifar10-1-noise-0.0005-0.0005/run-log-temp.txt', 'rb') as f:
    read_len = 0
    reader = csv.reader(f)
    for row in reader:
        if read_len < len(epochs):
            test_acc_noise_5.append(float(row[0]))
        read_len =read_len + 1
                
plt.clf()
#plt.yticks(np.arange(0,101))
plt.plot(epochs,test_acc,label='w.o noise')
plt.plot(epochs,test_acc_noise_1,label='w noise PROP:0.2, STD:0.0005')
plt.plot(epochs,test_acc_noise_2,linestyle = "-.",label='w noise PROP:0.1, STD:0.0005')
plt.plot(epochs,test_acc_noise_3,linestyle = ":",label='w noise PROP:0.05, STD:0.0005')
plt.plot(epochs,test_acc_noise_4,label='w noise PROP:0.005, STD:0.0005')
plt.plot(epochs,test_acc_noise_5,label='w noise PROP:0.0005, STD:0.0005')
plt.legend(loc='lower right', frameon=False)
plt.title("CIFAR10-Network 1 - Noise Impact on Test Accuracy")
plt.grid(linestyle = '-.')
plt.show()
```
